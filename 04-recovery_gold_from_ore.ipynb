### Вычисление эффективности обогащения
На обучающей выборке для признака `rougher.output.recovery` выполняем проверку, насколько правильно рассчитана эффективности обогащения чернового концентрата. 

В расчёте эффективности обогащения чернового концентрата используются следующие значения:
- `rougher.output.concentrate_au` - доля золота в концентрате после флотации/очистки
- `rougher.input.feed_au` - доля золота в сырье/концентрате до флотации/очистки
- `rougher.output.tail_au` - доля золота в отвальных хвостах после флотации/очистки

```python
# Вычисление эффективности обогащения
recovery_calculated = ((data_train['rougher.output.concentrate_au'] * (data_train['rougher.input.feed_au'] - data_train['rougher.output.tail_au'])) / 
                       (data_train['rougher.input.feed_au'] * (data_train['rougher.output.concentrate_au'] - data_train['rougher.output.tail_au']))) * 100
```

Покажем `MAE` между расчётами и значением признака эффективности обогащения. 

```python
# Вычисление средней абсолютной ошибки (MAE)
mean_absolute_error(data_train['rougher.output.recovery'], recovery_calculated)
```

Значение средней абсолютной ошибки `MAE` очень близка к нулю (_9.73512347450521e-15_). Это предполагает точность выполненных расчётов эффективности обогащения. Таким образом, можно с уверенностью сказать, что эффективность обогащения золота была рассчитана правильно на обучающей выборке.

В тестовой выборке отсутствуют следующие группы параметров:
- содержащие описание финальных характеристик (блок _этап_ в тестовой выборке не содержит `final`) продукта (тип параметра `output`)
- доля сырья после первичной очистки (этап `primary_cleaner` для типа параметра `output`)
- доля в сырье/концентрате после флотации/очистки (параметр продукта (т.е. ПОСЛЕ того, как сырьё прошло этап очистки) `output` для этапа флотации `rougher`)
- расчётные характеристики (тип параметра `calculation`) для флотации 
- доля продукта (тип параметра `output`) в отвальных хвостах (техпроцесс `tail`) после вторичной очистки (этап `secondary_cleaner`)

Таким образом в тестовой выборке отсутствуют все параметры, относящиеся к типам `output` и `calculation`, т.е. всё относящееся к данным после обработки (когда из сырья получен продукт) и расчётные характеристики (расчётные характеристики в исходных данных представлены только для этапа флотации).

### Создание функции для вычисления итоговой sMAPE
Определяем функцию `smape_function` для нахождения симметричного среднего абсолютного процентного отклонения `sMAPE` (Symmetric Mean Absolute Percentage Error). В данном случае пришлось писать полноценную функцию, в отличие от подсчёта [эффективности обогащения](#Вычисление-эффективности-обогащения), где достаточно было указать рассчётную формулу. Функция принимает на вход два аргумента: `actual` и `forecast` - фактические и прогнозные значения. Затем используем функцию для вычисления метрики sMAPE для данных из предварительно очищенной обучающей выборки.

Также необходимо обратить внимание на правило, которым будем руководствоваться далее. `sMAPE` выражается в процентах и интерпретируется следующим образом:

- меньшее значение sMAPE указывает на лучшую точность прогноза, а значение sMAPE, равное 0, указывает на идеальную точность прогноза;
- более высокое значение SMAPE указывает на худшую точность прогноза, а значение sMAPE, равное 100, указывает на совершенно неточный прогноз.

Таким образом, нужно спрогнозировать наболее низкое значение.

```python
def smape_function(actual, forecast):
    smape = 100 * (2 * abs(forecast - actual) / (abs(actual) + abs(forecast))).mean()
    return smape
```

Теперь определим `итоговую функцию sMAPE`

```python
def final_smape_function(rougher, final):
    final_smape = 0.25*rougher + 0.75*final
    return final_smape
```

### Подход к моделированию

Перед нами поставлена цель: _прогнозирование эффективности обогащения чернового концентрата `rougher.output.recovery` и эффективности обогащения финального концентрата `final.output.recovery`_. Целевые признаки количественные, а не категориальные. Стало быть, моделированием решается задача регрессии (не _классификации_). Для достижения цели созданы две функции (`smape_function` и `final_smape_function`), по значениям которых можно определить, насколько точно выполнено прогнозирование. 

1. Обученные модели необходимо проверить на правильность с помощью валидационной выборки. Ввиду отсутствия оного датасета мы предварительно выделим его из обучающей выборки `data_train_ml` в отношении 3:1.

2. Таким образом, построим 6 (шесть) модели прогнозирования:
    1. эффективность обогащения чернового концентрата с помощью дерева решений,
    2. эффективность обогащения чернового концентрата с помощью случайного леса,
    3. эффективность обогащения чернового концентрата с помощью линейной регрессии,
    4. эффективность обогащения финального концентрата с помощью дерева решений,
    5. эффективность обогащения финального концентрата с помощью случайного леса,
    6. эффективность обогащения финального концентрата с помощью линейной регрессии.

3. После предсказания выбираем лучшую модель на валидационной выборке. Аналогично после предсказания `final` также выбираем лучшую модель на валидационной выборке. 

4. После подсчёта `smape_function` [для лучших моделей на тестовой выборке](#Проверка-выбранной-модели-на-тестовой-выборке) из двух значений формируем `final_smape_function`. 

5. Выполним проверку модели на адекватность с помощью DummyRegression. Таков путь.

### Выделение валидационной выборки
Возможно нам сразу, ещё до этапа предобработки, нужно было выделить валидационную выборку, но сделаем это сейчас. 

```python
np.random.seed(42) # установка начального значения генератора случайных чисел в ходе обучения
```

```python
features = data_train_ml.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)         # объявление переменной признаков
target_rougher = data_train_ml['rougher.output.recovery']                                           # объявление целевого признака чернового концентрата
target_final = data_train_ml['final.output.recovery']                                               # объявление целевого признака финального концентрата
```

```python
# непосредственно разбиение на обучающую и валидационную выборки по признакам
features_train, features_valid, target_rougher_train, target_rougher_valid, target_final_train, target_final_valid = train_test_split(features, target_rougher, target_final, test_size=0.25)
```

### Составление прогноза с помощью модели дерева решений
#### Предсказание эффективности обогащения чернового концентрата

```python
# определение лучшей модели по smape
score_smape = make_scorer(smape_function, greater_is_better=False)
```

```python
%%time
# задаем сетку гиперпараметров для перебора GridSearchCV
param_grid = {
    'max_depth': range(1, 19),
    'min_samples_leaf': [1, 5, 10],
    'max_features': ['sqrt', 'log2', None]
}

# создаем объект модели регрессии дерева решений
model_tree_reg_rougher = DecisionTreeRegressor()

# создаем объект GridSearchCV
# оцениваем по neg_mean_absolute_error (чем ближе значение к нулю, тем лучше качество модели)
grid_search_tree_reg_rougher = GridSearchCV(
    model_tree_reg_rougher, 
    param_grid, 
    scoring=score_smape, 
    cv=5
)

# запускаем поиск оптимальных параметров
grid_search_tree_reg_rougher.fit(features_train, target_rougher_train)

# предсказания значения целевой переменной для валидационной выборки
valid_predictions_reg_rougher = grid_search_tree_reg_rougher.best_estimator_.predict(features_train)

# значение smape на валидационной выборке
result_reg_rougher = smape_function(target_rougher_train, valid_predictions_reg_rougher)
               
# выводим лучшие параметры и значение метрики
print(f'sMAPE эффективности обогащения чернового концентрата наилучшей модели на валидационной выборке {round(result_reg_rougher,3)}')
print(f'при следующих параметрах: max_depth={grid_search_tree_reg_rougher.best_params_["max_depth"]}, max_features=\'{grid_search_tree_reg_rougher.best_params_["max_features"]}\', min_samples_leaf={grid_search_tree_reg_rougher.best_params_["min_samples_leaf"]}')
print()
```

#### Предсказание эффективности обогащения финального концентрата
Рассчитываем прогноз по `sMAPE` для финального концентрата по модели дерева решений с использованием тех же подобранных значений гиперпараметров, что стали оптимальны в случае чернового концентрата. Возможно использовать нужно иные значения гиперпараметры, но так понимаю, первоначальные условия (значения гиперпараметров) должны быть идентичны.

```python
# сетка гиперпараметров для перебора GridSearchCV указана выше
# создаем объект модели регрессии дерева решений
model_tree_reg_final = DecisionTreeRegressor()

# создаем объект GridSearchCV
# оцениваем по neg_mean_absolute_error (чем ближе значение к нулю, тем лучше качество модели)
grid_search_tree_reg_final = GridSearchCV(
    model_tree_reg_final, 
    param_grid, 
    scoring=score_smape, 
    cv=5
)

# запускаем поиск оптимальных параметров
grid_search_tree_reg_final.fit(features_train, target_final_train)

# предсказания значения целевой переменной для валидационной выборки
valid_predictions_reg_final = grid_search_tree_reg_final.best_estimator_.predict(features_train)

# значение smape на валидационной выборке
result_reg_final = smape_function(target_final_train, valid_predictions_reg_final)
               
# выводим лучшие параметры и значение метрики
print(f'sMAPE эффективности обогащения финального концентрата наилучшей модели на валидационной выборке {round(result_reg_final,3)}')
print(f'при следующих параметрах: max_depth={grid_search_tree_reg_final.best_params_["max_depth"]}, max_features=\'{grid_search_tree_reg_final.best_params_["max_features"]}\', min_samples_leaf={grid_search_tree_reg_final.best_params_["min_samples_leaf"]}')
print()
```

### Составление прогноза с помощью модели случайного леса
#### Предсказание эффективности обогащения чернового концентрата

```python
%%time
# задаем сетку гиперпараметров для перебора GridSearchCV
param_grid_rfr = {
    'n_estimators': range(1,9,2),
    'max_depth': range(1,9,1),
    'min_samples_leaf': [1, 5, 10],
    'max_features': ['sqrt', 'log2', None]
}
    
# создаем объект модели случайного леса
model_rfr_reg_rougher = RandomForestRegressor()

# создаем объект GridSearchCV
# оцениваем по neg_mean_absolute_error (чем ближе значение к нулю, тем лучше качество модели)
grid_search_rfr_reg_rougher = GridSearchCV(
    model_rfr_reg_rougher, 
    param_grid_rfr, 
    scoring=score_smape, 
    cv=5
)

# запускаем поиск оптимальных параметров
grid_search_rfr_reg_rougher.fit(features_train, target_rougher_train)

# предсказания значения целевой переменной для валидационной выборки
valid_predictions_rfr_reg_rougher = grid_search_rfr_reg_rougher.best_estimator_.predict(features_train)

# значение smape на валидационной выборке
result_rfr_reg_rougher = smape_function(target_rougher_train, valid_predictions_rfr_reg_rougher)
               
# выводим лучшие параметры и значение метрики
print(f'sMAPE эффективности обогащения финального концентрата наилучшей модели на валидационной выборке {round(result_rfr_reg_rougher,3)}')
print(f'при следующих параметрах: n_estimators={grid_search_rfr_reg_rougher.best_params_["n_estimators"]}, max_depth={grid_search_rfr_reg_rougher.best_params_["max_depth"]}, max_features=\'{grid_search_rfr_reg_rougher.best_params_["max_features"]}\', min_samples_leaf={grid_search_rfr_reg_rougher.best_params_["min_samples_leaf"]}')
print()
```

```python
%%time
best_result_rfr_rougher = 10000
for est in tqdm(range(1,31,2)):
    for depth in range(1,9,1):
        
        # создание объекта модели регрессии случайный лес с двумя параметрами
        model_rfr_rougher = RandomForestRegressor(n_estimators=est, max_depth=depth)
        
        # обучение модели на тренировочной выборке
        model_rfr_rougher.fit(features_train, target_rougher_train) 
        
        # предсказания модели на валидационной выборке
        predictions_valid_rfr_rougher = model_rfr_rougher.predict(features_valid) 
        
        # значение smape на валидационной выборке
        result_rfr_rougher = smape_function(target_rougher_valid, predictions_valid_rfr_rougher)
        
        if result_rfr_rougher < best_result_rfr_rougher:
            best_model_rfr_rougher = model_rfr_rougher 
            best_result_rfr_rougher = result_rfr_rougher 
            best_est_rfr_rougher = est
            best_depth_rfr_rougher = depth

print(f'sMAPE эффективности обогащения чернового концентрата наилучшей модели на валидационной выборке: {round(best_result_rfr_rougher,3)} (при количестве деревьев = {best_est_rfr_rougher} и максимальной глубине = {best_depth_rfr_rougher})')
print()
```

#### Предсказание эффективности обогащения финального концентрата

```python
%%time
best_result_rfr_final = 10000
for est in tqdm(range(1,31,2)):
    for depth in range(1,9,1):
        
        # создание объекта модели регрессии случайный лес с двумя параметрами
        model_rfr_final = RandomForestRegressor(n_estimators=est, max_depth=depth)
        
        # обучение модели на тренировочной выборке
        model_rfr_final.fit(features_train, target_final_train) 
        
        # предсказания модели на валидационной выборке
        predictions_valid_rfr_final = model_rfr_final.predict(features_valid) 
        
        # значение smape на валидационной выборке
        result_rfr_final = smape_function(target_final_valid, predictions_valid_rfr_final)
        
        if result_rfr_final < best_result_rfr_final:
            best_model_rfr_final = model_rfr_final 
            best_result_rfr_final = result_rfr_final 
            best_est_rfr_final = est
            best_depth_rfr_final = depth

print(f'sMAPE эффективности обогащения финального концентрата наилучшей модели на валидационной выборке: {round(best_result_rfr_final,3)} (при количестве деревьев = {best_est_rfr_final} и максимальной глубине = {best_depth_rfr_final})')
print()
```

Обучение проведено для с помощью цикла перебора гиперпараметров `n_estimators` и `max_depth`, а также при подборе гиперпараметров через `GridSearchCV`. Результат, более близкий к желаемому, наблюдается во втором случае, однако перебрать большИй диапазон не представляется возможным, ввиду крайней длительности обучения модели в случае финального концентрата. Мне так ни разу не довелось дождаться выполнения расчёта. Поэтому блок выше закомментирован. 

В любом случае величина `sMAPE` у чернового концентрата почти в два раза ниже, чем в случае с финальным концентратом. Тоже самое, как и для модели дерева решений.

### Составление прогноза с помощью модели линейной регрессии
#### Предсказание эффективности обогащения чернового концентрата

```python
# создание объекта модели линейной регрессии без явного указания заданных гиперпараметров
model_linreg_rougher = LinearRegression()

# обучение модели для оценки эффективности обогащения чернового концентрата
model_linreg_rougher.fit(features_train, target_rougher_train)

# предсказания на валидационной выборке
predictions_linreg_valid_rougher = model_linreg_rougher.predict(features_valid)

# определение эффективности обогащения чернового концентрата на валидационной выборке в случае линейной регрессии
result_valid_linreg_rougher = smape_function(target_rougher_valid, predictions_linreg_valid_rougher)

print(f'sMAPE эффективности обогащения чернового концентрата наилучшей модели на валидационной выборке: {round(result_valid_linreg_rougher,2)}')
```

#### Предсказание эффективности обогащения финального концентрата

```python
# создание объекта модели линейной регрессии без явного указания заданных гиперпараметров
model_linreg_final = LinearRegression()

# обучение модели для оценки эффективности обогащения чернового концентрата
model_linreg_final.fit(features_train, target_final_train)

# предсказания на валидационной выборке
predictions_linreg_valid_final = model_linreg_final.predict(features_valid)

# определение эффективности обогащения финального концентрата на валидационной выборке в случае линейной регрессии
result_valid_linreg_final = smape_function(target_final_valid, predictions_linreg_valid_final)

print(f'sMAPE эффективности обогащения финального концентрата наилучшей модели на валидационной выборке: {round(result_valid_linreg_final,2)}')
```

Проверку выполним для каждого из признаков по отдельности. Ввиду того, что обучение модели дерева решений `model_tree_reg_` происходило с применением `GridSearchCV`, наименование тестируемой модели также иное.

```python
# проверка модели дерева решений для показателя эффективности обогащения чернового концентрата
features_test_rougher = data_test_tar['rougher.output.recovery']
test_predictions_rougher = grid_search_rfr_reg_rougher.predict(features_test)
print(f'sMAPE чернового концентрата = {smape_function(test_predictions_rougher, targer_test_rougher)}')

# проверка модели дерева решений для показателя эффективности обогащения чернового концентрата
features_test_final = data_test_tar['final.output.recovery']
test_predictions_final = grid_search_tree_reg_final.predict(features_test)
print(f'sMAPE финального концентрата = {smape_function(test_predictions_final, targer_test_final)}')
```

## Проверка модели на адекватность

```python
model_dummy_r = DummyRegressor(strategy='mean')
model_dummy_r.fit(features_train, target_rougher_train)
target_pred_r = model_dummy_r.predict(features_train)
rougher_smape_ = smape_function(target_rougher_train, target_pred_r)
print('sMAPE эффективности обогащения чернового концентрата на предсказательной модели DummyRegressor', rougher_smape_)
```

```python
model_dummy = DummyRegressor(strategy='mean')
model_dummy.fit(features_train, target_final_train)
target_pred = model_dummy.predict(features_train)
final_smape_ = smape_function(target_final_train, target_pred)
print('sMAPE эффективности обогащения финального концентрата на предсказательной модели DummyRegressor', final_smape_)
```

Результат показал, что модель на адекватность возвращает значение `sMAPE` больше, чем расчёты всех обученных моделей и для финального концентрата, и, гораздо лучше, для чернового концентрата. Очень хорошо.
