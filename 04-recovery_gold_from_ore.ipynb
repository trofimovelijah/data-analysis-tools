## Вычисление эффективности обогащения
На обучающей выборке для признака `rougher.output.recovery` выполняем проверку, насколько правильно рассчитана эффективности обогащения чернового концентрата. 

В расчёте эффективности обогащения чернового концентрата используются следующие значения:
- `rougher.output.concentrate_au` - доля золота в концентрате после флотации/очистки
- `rougher.input.feed_au` - доля золота в сырье/концентрате до флотации/очистки
- `rougher.output.tail_au` - доля золота в отвальных хвостах после флотации/очистки

```python
# Вычисление эффективности обогащения
recovery_calculated = ((data_train['rougher.output.concentrate_au'] * (data_train['rougher.input.feed_au'] - data_train['rougher.output.tail_au'])) / 
                       (data_train['rougher.input.feed_au'] * (data_train['rougher.output.concentrate_au'] - data_train['rougher.output.tail_au']))) * 100
```

Покажем `MAE` между расчётами и значением признака эффективности обогащения. 

```python
# Вычисление средней абсолютной ошибки (MAE)
mean_absolute_error(data_train['rougher.output.recovery'], recovery_calculated)
```

Значение средней абсолютной ошибки `MAE` очень близка к нулю (_9.73512347450521e-15_). Это предполагает точность выполненных расчётов эффективности обогащения. Таким образом, можно с уверенностью сказать, что эффективность обогащения золота была рассчитана правильно на обучающей выборке.

В тестовой выборке отсутствуют следующие группы параметров:
- содержащие описание финальных характеристик (блок _этап_ в тестовой выборке не содержит `final`) продукта (тип параметра `output`)
- доля сырья после первичной очистки (этап `primary_cleaner` для типа параметра `output`)
- доля в сырье/концентрате после флотации/очистки (параметр продукта (т.е. ПОСЛЕ того, как сырьё прошло этап очистки) `output` для этапа флотации `rougher`)
- расчётные характеристики (тип параметра `calculation`) для флотации 
- доля продукта (тип параметра `output`) в отвальных хвостах (техпроцесс `tail`) после вторичной очистки (этап `secondary_cleaner`)

Таким образом в тестовой выборке отсутствуют все параметры, относящиеся к типам `output` и `calculation`, т.е. всё относящееся к данным после обработки (когда из сырья получен продукт) и расчётные характеристики (расчётные характеристики в исходных данных представлены только для этапа флотации).

## Создание функции для вычисления итоговой sMAPE
Определяем функцию `smape_function` для нахождения симметричного среднего абсолютного процентного отклонения `sMAPE` (Symmetric Mean Absolute Percentage Error). В данном случае пришлось писать полноценную функцию, в отличие от подсчёта [эффективности обогащения](#Вычисление-эффективности-обогащения), где достаточно было указать рассчётную формулу. Функция принимает на вход два аргумента: `actual` и `forecast` - фактические и прогнозные значения. Затем используем функцию для вычисления метрики sMAPE для данных из предварительно очищенной обучающей выборки.

```python
def smape_function(actual, forecast):
    smape = 100/len(actual) * sum(2 * abs(forecast - actual) / (abs(actual) + abs(forecast)))
    return smape
```

Теперь определим `итоговую функцию sMAPE`

```python
def final_smape_function(rougher, final):
    final_smape = 0.25*rougher + 0.75*final
    return final_smape
```
