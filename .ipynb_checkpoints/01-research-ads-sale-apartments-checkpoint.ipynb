<h4>Изменение типов данных для ряда параметров

Для предварительной оценки содержимого столбцов вновь выведем первые 5 строк датасета на экран, а также информацию по датасету.

На основе указанных данных нуждаются в обязательной замене:
 - `first_day_exposition` - дата публикации:
  - _object_ на **datetime** - <span style="color:blue">дата указа в формате `2019-03-07T00:00:00`. Для времени есть отдельный тип данных. При этом характерно, что время суток не указано ни разу.
 - `balcony` - число балконов: 
  - _float64_ на **int64** - <span style="color:blue">даже самый маленький и несуразный балкон уже считается за целую единицу. Использование вещественного типа данных излишне.
 - `days_exposition` - сколько дней было размещено объявление (от публикации до снятия):
  - _float64_ на **int64** - <span style="color:blue">указывается только целое число дней. Таким образом, вещественный тип данных излишен.
 - `last_price` - цена на момент снятия с публикации:
  - _float64_ на **int64** - <span style="color:blue">цена может обозначаться в виде дробных значений. Но в нашем случае речь идёт о миллионах, поэтому смело значения параметра можно округлять до целого, если вдруг в цене жилья упоминаются копейки.
 - `parks_around3000` - число парков в радиусе 3 км:
  - _float64_ на **int64** - <span style="color:blue">парки могут быть только целые, без дробей
 - `ponds_around3000` - число водоёмов в радиусе 3 км:
  - _float64_ на **int64** - <span style="color:blue">водоёмы, как и парки, могут быть только целые, без дробей
 - `floors_total` - всего этажей в доме:
  - _float64_ на **int64** - <span style="color:blue">в жизни встречаются полуторные этажи, например, мансарды и технические этажи в производственных зданиях. Но на практике не доводилось встречать в объявлениях где-либо указание нецелочисленных этажей. Поскольку у нас уже есть параметр `floor` с типом данных `int64`, то не подразумевается, что конкретный этаж указан в дробном виде. Следовательно не подразумевается использование и переменной`floor_total` в дробном виде. Хотя можно предположить, что верно и обратное - приведение параметра `floor` к типу данных `float64`. Но мы сойдёмся на первом варианте

```python
try:
    data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])
    data['balcony'] = data['balcony'].astype('int64')
    data['days_exposition'] = data['days_exposition'].astype('int64')
    data['parks_around3000'] = data['parks_around3000'].astype('int64')
    data['ponds_around3000'] = data['ponds_around3000'].astype('int64')
    data['last_price'] = data['last_price'].astype('int64')
    data['floors_total'] = data['floors_total'].astype('int64')
except:
    print(f'ещё раз посмотри вывод данных команды `data.head()`\n {data.head()}')
```

У следующих параметров также возможно поменять тип данных, хотя и необязательно:
 - `cityCenters_nearest` — расстояние до центра города (м):
  - _float64_ на **int64** - <span style="color:blue">медианное значение расстояния до центра города более 11 километров, поэтому использование дробей не имеет практического смысла
 - `parks_nearest` - расстояние до ближайшего парка (м):
  - _float64_ на **int64** - <span style="color:blue">крайне сложно предположить случай, когда замеряют расстояние до географических объектов с точностью до десятых дробей
 - `ponds_nearest` - расстояние до ближайшего водоёма (м):
  - _float64_ на **int64** - <span style="color:blue">тоже самое, что и для расстояния до ближайшего парка

Но мы всё же поменяем

```python
try:
    data['cityCenters_nearest'] = data['cityCenters_nearest'].astype('int64')
    data['parks_nearest'] = data['parks_nearest'].astype('int64')
    data['ponds_nearest'] = data['ponds_nearest'].astype('int64')
except:
    print('если выпадет ошибка, то и ладно. Это ж необязательно')
```

```python
data['locality_name_uniform'] = data['locality_name'].str.replace('поселок |городской поселок |коттеджный поселок |коттеджный посёлок |коттеджный пгт |поселок городского типа |городского типа |городской |пгт |посёлок посёлок ', 'посёлок ', regex=True)
data['locality_name_uniform'] = data['locality_name_uniform'].str.replace('поселок при железнодорожной станции |посёлок при железнодорожной станции |поселок станции', 'посёлок при станции ', regex=True)
data['locality_name_uniform'] = data['locality_name_uniform'].str.replace('садовое товарищество |садоводческое некоммерческое товарищество ', 'садовое товарищество ', regex=True)
data['locality_name_uniform'] = data['locality_name_uniform'].str.replace('посёлок посёлок ', 'посёлок ', regex=True)
```

```python
data['locality_name_uniform'].value_counts().count()
```

Таким образом мы убрали все неявные дубликаты, создав отдельный столбец `locality_name_uniform`. Теперь в таблице присутствуют лишь города и несколько типов населённых пунктов, которые решено было оставить.

<h4>Определение средней цены одного квадратного метра в 10 н.п.

Средняя цена одного кв.м. в 10 н.п. с наибольшим числом объявлений

Порядок действий:
- сгруппируем данные по населённым пунктам. 
- вычислим количество объявлений в каждом населённом пункте с сортировкой по убыванию
- выбирем 10 населённых пунктов с наибольшим числом объявлений
- вычислим среднюю цену одного квадратного метра для каждого населённого пункта

```python
grouped_locality_name = data.groupby('locality_name')

counted_locality_name = grouped_locality_name['price_one_square'].count().sort_values(ascending=False)

top_localities = counted_locality_name.head(10).index

filtered_data = data[data['locality_name'].isin(top_localities)]

mean_price_one_square = filtered_data.groupby('locality_name')['price_one_square'].mean()

mean_price_one_square
```

Н.п. с самой высокой и самой низкой стоимостью кв.м.

```python
highest_price_locality = mean_price_one_square.idxmax()
lowest_price_locality = mean_price_one_square.idxmin()

print(f'Населённый пункт с самой высокой стоимостью квадратного метра: {highest_price_locality}')
print(f'Населённый пункт с самой низкой стоимостью квадратного метра: {lowest_price_locality}')
```

Также можно заметить, что для квартир цены начинают резко увеличиваться на расстоянии около 7-8 км от центра, а для домов - на расстоянии около 20-25 км.
